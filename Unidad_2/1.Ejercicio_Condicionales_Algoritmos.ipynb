📌 Consigna:
 Generar un algoritmo para indicar si el usuario tiene la edad necesaria para conducir.
1. Crear una variable para guardar la edad del usuario con input() y convertirla a número con int().
2. Crear una estructura condicional if que evalúe si la edad es mayor o igual a 16.

2️⃣ Actividad 2: Helado 🍦
📌 Consigna:
 Generar un algoritmo para consultar si hay o no un gusto de helado disponible.
1. Crear una variable que guarde el gusto ingresado por el usuario con input().
2. Evaluar con if si el gusto ingresado coincide con alguno de estos: "chocolate", "vainilla", "frutilla" o "dulce de leche".
3. Si está disponible, mostrar: "Sí, hay ____".
4. Si no, mostrar: "No hay ____".


```python
# 1
gusto = input('Ingresa el sabor del helado: ').lower()
# 2
if gusto in ['chocolate', 'vainilla', 'frutilla', 'dulce de leche']:
  print(f'Si hay {gusto}')
else:
  print(f'No hay {gusto}')
```

3️⃣ Actividad 3: Comparador de números 🔢
📌 Consigna:
 Crear un algoritmo que compare cuatro números ingresados por el usuario y muestre cuál es mayor.
1. Pedir al usuario que ingrese cuatro números (convertidos a int).
2. Usar condicionales if, elif y else para:
    - Mostrar si el primer número es mayor.
   - Mostrar si el segundo número es mayor.
   - Mostrar si el tercer número es mayor.
   - Mostrar si el cuarto número es mayor.
    - O si todos son iguales.

```python
num_1 = int(input('Ingresa el primer número: '))
num_2 = int(input('Ingresa el segundo número: '))
num_3 = int(input('Ingresa el tercero número: '))
num_4 = int(input('Ingresa el cuarto número: '))

if num_1 > num_2 & num_1 > num_3 & num_1 > num_4:
  print(f'El número {num_1} es el mayor')
elif num_2 > num_1 & num_2 > num_3 & num_4:
  print(f'El número {num_2} es el mayor')
elif num_3 > num_1 & num_3 > num_2 & num_3 > num_4:
  print(f'El número {num_3} es el mayor')
elif num_4 > num_1 & num_4 > num_2 & num_4 > num_3:
  print(f'El número {num_3} es el mayor')
else:
  print('Son todos los números iguales')
```

4️⃣Actividad 4: Condición de estudiante 🧑‍🎓
📌 Consigna: Generar un algoritmo para cargar por teclado las notas obtenidas por un estudiante en tres parciales rendidos durante el cursado de una asignatura universitaria (si no hizo el parcial deberá cargarse 0). Además, se carga la nota final que ese estudiante obtuvo en el desarrollo de los trabajos prácticos en esa misma asignatura. Se sabe que al terminar el cursado de la materia, todo alumno puede quedar en uno de los siguientes estados académicos.
1. Libre: si no llegó a cumplir con las condiciones para ser Regular.
2. Regular: todos los parciales rendidos, al menos dos de los tres parciales con nota de 4 (cuatro) o más y además obtuvo nota de 4 (cuatro) o más en la nota final de trabajos prácticos.
3. Promocionado: si aprobó los tres parciales con nota de 7 (siete) o más pero con promedio entre ellos de 9 (nueve) o más, y además obtuvo nota de 8 o más en la nota final del práctico.
4. Aprobado: si aprobó los tres parciales con nota de 7 (siete) o más pero con promedio entre ellos de 8 (ocho) o más, y además obtuvo nota de 8 (ocho) o más en la nota final del práctico.
El programa debe determinar y mostrar por pantalla el estado en que finalmente quedó el estudiante.


```python
p1 = float(input('Ingresa tu primera nota (si no lo hiciste es 0): '))
p2 = float(input('Ingresa tu segunda nota (si no lo hiciste es 0): '))
p3 = float(input('Ingresa tu tercera nota (si no lo hiciste es 0): '))
tp = float(input('Ingresa la nota final de los trabajos prácticos: '))

promedio = (p1 + p2 + p3) / 3
parciales_rendidos = (p1 > 0) and (p2 > 0) and (p3 > 0)
parciales_7 = (p1 >= 7) and (p2 >= 7) and (p3 >= 7)

if parciales_7 and promedio >= 9 and tp >= 8:
  estado = 'Promocionado'
elif parciales_7 and promedio >= 8 and tp >= 8:
  estado = 'Aprobado'
elif parciales_rendidos and ((p1 >= 4 and p2 >= 4) or (p1 >= 4 and p3 >= 4) or (p2 >= 4 and p3 >= 4)) and tp >= 4:
  estado = 'Regular'
else:
  estado = 'Libre'
print(f'El estado del estudianto es {estado}')
```
