4️⃣Actividad 4: Condición de estudiante 🧑‍🎓
📌 Consigna: Generar un algoritmo para cargar por teclado las notas obtenidas por un estudiante en tres parciales rendidos durante el cursado de una asignatura universitaria (si no hizo el parcial deberá cargarse 0). Además, se carga la nota final que ese estudiante obtuvo en el desarrollo de los trabajos prácticos en esa misma asignatura. Se sabe que al terminar el cursado de la materia, todo alumno puede quedar en uno de los siguientes estados académicos.
1. Libre: si no llegó a cumplir con las condiciones para ser Regular.
2. Regular: todos los parciales rendidos, al menos dos de los tres parciales con nota de 4 (cuatro) o más y además obtuvo nota de 4 (cuatro) o más en la nota final de trabajos prácticos.
3. Promocionado: si aprobó los tres parciales con nota de 7 (siete) o más pero con promedio entre ellos de 9 (nueve) o más, y además obtuvo nota de 8 o más en la nota final del práctico.
4. Aprobado: si aprobó los tres parciales con nota de 7 (siete) o más pero con promedio entre ellos de 8 (ocho) o más, y además obtuvo nota de 8 (ocho) o más en la nota final del práctico.
El programa debe determinar y mostrar por pantalla el estado en que finalmente quedó el estudiante.


```python
p1 = float(input('Ingresa tu primera nota (si no lo hiciste es 0): '))
p2 = float(input('Ingresa tu segunda nota (si no lo hiciste es 0): '))
p3 = float(input('Ingresa tu tercera nota (si no lo hiciste es 0): '))
tp = float(input('Ingresa la nota final de los trabajos prácticos: '))

promedio = (p1 + p2 + p3) / 3
parciales_rendidos = (p1 > 0) and (p2 > 0) and (p3 > 0)
parciales_7 = (p1 >= 7) and (p2 >= 7) and (p3 >= 7)

if parciales_7 and promedio >= 9 and tp >= 8:
  estado = 'Promocionado'
elif parciales_7 and promedio >= 8 and tp >= 8:
  estado = 'Aprobado'
elif parciales_rendidos and ((p1 >= 4 and p2 >= 4) or (p1 >= 4 and p3 >= 4) or (p2 >= 4 and p3 >= 4)) and tp >= 4:
  estado = 'Regular'
else:
  estado = 'Libre'
print(f'El estado del estudianto es {estado}')
```

📌 Consigna: Generar un algoritmo para una empresa que paga a sus vendedores un sueldo bruto igual a la suma de un sueldo básico de $550.000 más una comisión que es igual a un porcentaje del monto total vendido. El porcentaje por comisión depende de la categoría del vendedor de acuerdo a la siguiente tabla:

Por otro lado, si el sueldo bruto del vendedor es mayor a $300.000, se efectúa un descuento igual al 15% del sueldo bruto; en caso contrario, se efectúa un descuento igual al 10% del sueldo bruto.
Diseñe un algoritmo que determine el sueldo básico, la comisión, el sueldo bruto, el descuento y el sueldo neto de un vendedor de la empresa.



```python
sueldobase = 550000
categoria = str(input('Ingrese su categoría: ')).upper()
ventas = int(input('Ingrese el total de ventas: '))

if categoria == 'SENIOR':
  comision = 0.1425 * ventas
elif categoria == 'SEMI-SENIOR':
  comision = 0.13 * ventas
elif categoria == 'JUNIOR':
  comision = 0.1175 * ventas
else:
  print('Ingesa una categoría válida')

sueldobruto = sueldobase + comision

if sueldobruto > 300000:
  descuento = 0.15 * sueldobruto
  sueldoneto =  sueldobruto - descuento
else:
  descuento = 0.1 * sueldobruto
  sueldoneto = sueldobruto - descuento

print(f'Tu sueldo básico es: {sueldobase}')
print(f'Tu comisión es: {comision}')
print(f'Tu sueldo bruto es: {sueldobruto}')
print(f'Tu descuento es : {descuento}')
print(f'Tu sueldo neto es: {sueldoneto}')
```

🔟Actividad 10: Asociación de vitivinicultores 🍇🍷
📌 Consigna: Generar un programa para una asociación de vinicultores que tiene como política fijar un precio inicial al kilo de uva, la cual se clasifica en tipos A y B, y además en tamaños 1 y 2. Cuando se realiza la venta del producto, ésta es de un solo tipo y tamaño, se requiere determinar cuánto recibirá un productor por la uva que entrega en un embarque, considerando lo siguiente: si es de tipo A, se le cargan 200 al precio inicial cuando es de tamaño 1; y 300 si es de tamaño 2. Si es de tipo B, se rebajan 300 cuando es de tamaño 1, y 500 cuando es de tamaño 2. Al finalizar de procesar los datos debe determinar la ganancia obtenida y mostrar el resultado.


```python
tipo_u = input('Tipo de la uva A/B: ').upper()
tam_u = int(input('Tamaño de la uva 1/2: '))
kilo_u = float(input('ingresa la cantidad de kilos: '))
precio_ini = float(input('ingresar el precio kg de uva: '))

if tipo_u == 'A':
  if tam_u == 1:
    precio_t = precio_ini + 200
  elif tam_u == 2:
    precio_t = precio_ini + 300
  else:
    print('Tamaño de uva no válido')
elif tipo_u == 'B':
  if tam_u == 1:
    precio_t = precio_ini - 300
  elif tam_u == 2:
    precio_t = precio_ini - 500
  else:
    print('Tamaño de uva no válido')
else:
  print('Tipo de uva no válido')

ganancia = precio_t * kilo_u
print(f'La ganancia total obtenida es ${ganancia}')
```
