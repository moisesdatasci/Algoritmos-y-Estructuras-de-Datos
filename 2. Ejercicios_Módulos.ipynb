```python
from datetime import datetime
```

```python
ahora = datetime.now()
print(f'Fecha y hora actual: {ahora}')
print(f'Año: {ahora.year}')
print(f'Mes: {ahora.month}')
print(f'Día: {ahora.day}')
print(f'Día de la semana: {ahora.weekday()}')
```

2️⃣ Actividad con time

📌 Consigna:

1. Importa el módulo time.
2. Muestra la hora actual en pantalla en formato H:M:S.
3. Espera 3 segundos.
4. Muestra el mensaje "¡Tiempo finalizado!".


```python
import time
```

```python
hora_actual = time.strftime('%H:%M:%S')
print(f"Hora actual: {hora_actual}")

print('Esperando 3 sengundos...')
time.sleep(3)
print('¡Tiempo finalizado!')

hora_final = time.strftime('%H:%M:%S')
print(f"Hora final: {hora_final}")
```
